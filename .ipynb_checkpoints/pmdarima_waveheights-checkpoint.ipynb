```python
import multiprocessing
import os
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import pickle
from joblib import Parallel, delayed
```

```python
arima_train = pd.to_numeric(arima_train, errors='coerce')
arima_test = pd.to_numeric(arima_test, errors='coerce')
#print(arima_train.dtypes)
#print(arima_test.dtypes)
```

```python
# Split the test data into chunks
num_chunks = 10  # Define the number of chunks you want to split into
chunk_size = len(arima_test) // num_chunks
chunks = [arima_test[i:i + chunk_size] for i in range(0, len(arima_test), chunk_size)]
```

```python
p, d, q = 4, 1, 1

def fit_arima_chunk(history, test_chunk):
    predictions = []
    history = list(history)
    for t in range(len(test_chunk)):
        try:
            model = ARIMA(history, order=(p, d, q))
            model_fit = model.fit()
            output = model_fit.forecast()
            yhat = output[0]
            predictions.append(yhat)
            obs = test_chunk[t]
            history.append(obs)
        except Exception as e:
            print(f'Error fitting model at step {t}: {e}')
            predictions.append(np.nan)  # Handle error by appending NaN
    return predictions, history
```

```python
# Use joblib to parallelize the process
num_cores = 28 
results = Parallel(n_jobs=num_cores)(delayed(fit_arima_chunk)(arima_train, chunk) for chunk in chunks)
```

```python
# Collect and merge results
final_history = list(arima_train)
all_predictions = []
for predictions, updated_history in results:
    all_predictions.extend(predictions)
    final_history.extend(updated_history[len(arima_train):])  # Merge updated history without duplicating initial history
```

```python
#view results
all_predictions
```

```python
# Function to fit ARIMA model on the entire dataset and save it
def fit_and_save_arima_model(data, order, model_path):
    try:
        model = ARIMA(data, order=order)
        model_fit = model.fit()
        
        # Save the model to disk
        with open(model_path, 'wb') as pkl_file:
            pickle.dump(model_fit, pkl_file)
            
        print(f'Model saved to {model_path}')
    except Exception as e:
        print(f'Error fitting and saving model: {e}')
```

```python
# Fit on the entire cumulative history and save the model
fit_and_save_arima_model(final_history, (p, d, q), 'arima_model.pkl')
```

```python
# Function to load the ARIMA model
def load_arima_model(model_path):
    try:
        with open(model_path, 'rb') as pkl_file:
            model_fit = pickle.load(pkl_file)
        print(f'Model loaded from {model_path}')
        return model_fit
    except Exception as e:
        print(f'Error loading model: {e}')
        return None
```

```python
# Loading the Model
model_path = 'arima_model.pkl'
loaded_model = load_arima_model(model_path)
```
